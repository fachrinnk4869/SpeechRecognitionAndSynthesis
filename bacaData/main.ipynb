```python
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
#Membaca data dari file, y: buffer, sr: sampling rate
y,sr = librosa.load("aiueo.wav")
#tampilkan tipe data nya
print(type(y),type(sr))
```

```python

#tampilkan dimensi datanya dan nilai sr nya
print(y.shape, sr)
```

```python
#tampilkan waveform
librosa.display.waveshow(y,sr=sr)
plt.show()
```

```python
#durasi dalam detik
print(librosa.samples_to_time(len(y), sr=sr))
```
